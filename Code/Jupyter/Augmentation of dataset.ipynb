# Augmentation of dataset

```python
import augmolino.augmentation as augmolino
import os

user_mode = True
```

```python
# Datasets/ESC50_augmented/
if user_mode:
    main_folder = input("Enter path to main folder: ")
else:
    main_folder = "Datasets/UrbanSound8k_augmented/"
```

```python
print("Starting augmentation, this may take some time...")
d = input("Are you sure you want to proceed? y/n: ")
i = 0
n_augmentations = 5
noise_fp = "demo_data/noise_b_mili.wav"
if d == "y":

    path, dirs, files = next(os.walk(main_folder))
    print("\n\nAugmenting...")
    print(f'selected path: {path}')
    dirs_format = '\n\t'.join(dirs)
    print(f'registered dirs: \n\t{dirs_format}')
    for nscls in dirs:
        print(f'\nAugmenting in Folder < {nscls} >:')
        path, dirs, files = next(os.walk(main_folder + nscls + '/'))
        for wav in files:
            fp = main_folder + nscls + '/' + wav
            print(f'\tfile: {wav}')

            fp_dest_psu1 = main_folder + nscls + '/psu1_' + wav
            fp_dest_psu2 = main_folder + nscls + '/psu2_' + wav
            fp_dest_psd1 = main_folder + nscls + '/psd1_' + wav
            fp_dest_psd2 = main_folder + nscls + '/psd2_' + wav

            fp_dest_ts1 = main_folder + nscls + '/ts1_' + wav
            fp_dest_ts2 = main_folder + nscls + '/ts2_' + wav
            fp_dest_ts3 = main_folder + nscls + '/ts3_' + wav
            fp_dest_ts4 = main_folder + nscls + '/ts4_' + wav

            fp_dest_offs05 = main_folder + nscls + '/offs05_' + wav
            fp_dest_offs10 = main_folder + nscls + '/offs10_' + wav
            fp_dest_offs15 = main_folder + nscls + '/offs15_' + wav
            fp_dest_offs20 = main_folder + nscls + '/offs20_' + wav
            fp_dest_offs25 = main_folder + nscls + '/offs25_' + wav
            fp_dest_offs30 = main_folder + nscls + '/offs30_' + wav
            fp_dest_offs35 = main_folder + nscls + '/offs35_' + wav

            fp_dest_n = main_folder + nscls + '/n_' + wav

            i += 1

            
            # Enter augmentation methods here:
            #---------------------------------
            
            '''
            As proposed by Justin Salamon & Juan Pablo Bello in "Deep Convolutional Neural Networks and Data
            Augmentation for Environmental Sound Classification" it seems that "pitch shift 1 & 2", "back
            ground noise" and "time stretch" yield the best increase in performance.

            Also, since the framesize is fixed and no hopping is introduced yet, the files should also be
            offset to simulate sounds which cross multiple frames (due to their lenght or start-time)
            '''
            
            augmolino.pitchShift(fp, fp_dest_psu1, 1)
            augmolino.pitchShift(fp, fp_dest_psu2, 2)
            augmolino.pitchShift(fp, fp_dest_psd1, -1)
            augmolino.pitchShift(fp, fp_dest_psd2, -2)

            augmolino.timeStretch(fp, fp_dest_ts1, 0.81)
            augmolino.timeStretch(fp, fp_dest_ts2, 0.93)
            augmolino.timeStretch(fp, fp_dest_ts3, 1.07)
            augmolino.timeStretch(fp, fp_dest_ts4, 1.23)
            
            augmolino.offsetAudio(fp, fp_dest_offs05, -0.5)
            augmolino.offsetAudio(fp, fp_dest_offs10, -1.0)
            augmolino.offsetAudio(fp, fp_dest_offs15, -1.5)
            augmolino.offsetAudio(fp, fp_dest_offs20, -2.0)
            augmolino.offsetAudio(fp, fp_dest_offs25, -2.5)
            augmolino.offsetAudio(fp, fp_dest_offs30, -3.0)
            augmolino.offsetAudio(fp, fp_dest_offs35, -3.5)

            augmolino.fuseAudio(fp, noise_fp, fp_dest_n, lvl_ratio=0.5)
            
            #---------------------------------
        print(f'Folder < {nscls} > done.')
    print(f'\n\naugmentation done. {i} source files augmented to {i * n_augmentations} new files. Total file count: {i * (n_augmentations + 1)}')   
else:
    print("op cancelled.")
```

```python
import numpy as np
import octafilt3r.plot as o3p
import octafilt3r.weightings as o3w
```

```python
fs1 = 32000
fs2 = 48000

sos_cor, sos_real = o3w._ICS43432_correction_sos(fs=fs1)
print(sos_cor)
o3p._display_filt(sos_real, fs1, "ICS43432-frequency response")
o3p._display_filt(sos_cor, fs1, "ICS43432-correction curve")

sos_cor, sos_real = o3w._ICS43432_correction_sos(fs=fs2)
print(sos_cor)
o3p._display_filt(sos_real, fs2, "ICS43432-frequency response")
o3p._display_filt(sos_cor, fs2, "ICS43432-correction curve")
```

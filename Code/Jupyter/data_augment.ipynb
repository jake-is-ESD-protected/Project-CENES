```python
# needed imports
import librosa as lr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import librosa.display
import IPython.display as ipd
import soundfile as sf
from IPython.core.display import display

def timeStretch(fp_source, fp_dest, factor):
    x, sr = lr.load(fp_source)
    x_new = lr.effects.time_stretch(x, factor)
    sf.write(fp_dest, x_new, sr)
    return fp_dest

path = 'UrbanSound8k/audio/fold1/'
f_name = '7061-6-0-0.wav'
fp = path + f_name
fp_new = 'augmented_data/' + 'augm_' + f_name
print(fp_new)

timeStretch(fp, fp_new, 0.9)

display(ipd.Audio(fp_new))
```

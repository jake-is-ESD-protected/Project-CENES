```python
import octafilt3r.plot as o3p
import octafilt3r.weightings as o3w
import pipeline.config as conf

param_dict = conf.open_params()

fs = param_dict["fs"]
sos = o3w._A_sos(fs)
print(sos)
o3p._display_filt(sos, fs, "A-weighting")

for stage in sos:
    for coeff in stage:
        print(f'{coeff:.12f}')
```
